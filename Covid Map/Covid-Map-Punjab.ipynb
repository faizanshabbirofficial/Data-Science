```python
import folium as fo
import pandas as pd
```

```python
punjab_map = fo.Map(location = [30.85, 72.31], zoom_start = 7, tiles = 'OpenStreetMap')
```

```python
punjab_map
```

```python
df = pd.read_csv("Covid-Data.csv")
df
```

```python
total = 0
for cases in df['Total Cases']:
    total = int(cases) + int(total)
```

```python
for city, lat, long, tc, de, re, ac in zip(list(df['City']), list(df['Latitude']), list(df['Longitude']), list(df['Total Cases']), list(df['Deaths']), list(df['Recoveries']), list(df['Active Cases'])):
    fo.CircleMarker(location = [lat,long], radius = (tc/total)*100, color='red', fill = False, fill_color = 'blue').add_to(punjab_map)
    htmlData = 'Total Cases: ' + str(tc) + '' + 'Deaths: '+ str(de) + '' + 'Recoveries: ' + str(re) + '' + 'Active Cases: ' + str(ac)
    iframe = fo.IFrame(htmlData, width=200, height=250)
    popupData = fo.Popup(iframe, max_width=900)
    fo.Marker(location = [lat,long], popup=popupData).add_to(punjab_map)
```
